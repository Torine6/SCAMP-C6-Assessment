```python
fakejob_postings = fakejob_data.groupby('industry')[['industry','fraudulent']].sum().reset_index()   
```

```python
fakejob_postings             #grouping the data by number of fraudulent cases
```

```python
fake_jobs = fakejob_postings[fakejob_postings['fraudulent']>0]
```

```python
fake_jobs          #industries with at least(>=) one fake job posting
```

```python
import matplotlib.pyplot as plt    #graphical display of the results
```

```python
bargraph = fake_jobs.plot.bar(x='industry',y='fraudulent',title='Graph of fake job postings for different industries',width=3,fontsize=7)
```
